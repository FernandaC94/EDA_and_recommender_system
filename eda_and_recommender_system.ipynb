## 5. Recommender System 
In this section we will try to build a basic recomender system, although we have a small dataset. This section will focus more in learning how recommender system work rather than actually build an accurate recommender system. We will be building more robus ones with a larger dataset

Reading suggestion: https://developers.google.com/machine-learning/recommendation/content-based/summary 
There are three types of recommender system:
* content based filtering: check similarity between the products or users. This type of recommender system handdle the problem known as 'cold start'(we dont have any data about the user) 

* Collaborative filtering: check similarities between products and other features, such as past records of the user or past records of similar users.

* Hybrids filtering: Use a combination of content and collaborative filters.

### 5.1 Content based filtering based on description of the product (product_description) and category

 we could also use the about_product column to do our analysis but we decided to keep it simple. 

#### Some data preprocesing

```python
# we are going to take the description of the product 
#and the category column concatenate it together to conver it in
#only one feature that we will be using for our model 
data=df['product_description']+ ' '+ df['category']

def cleaning(txt):
    
    # we clean the special characters in our data
    txt=re.sub('([^A-Za-z])',' ',txt)
    txt=re.sub('\s+',' ',txt).lower()
    #lemmatization initializer
    wnl=WordNetLemmatizer()
    #list of stop words
    stop_words=set(stopwords.words('english'))
    #lemmatization and stop words removed. Including only words with length grater than 2
    txt=' '.join([wnl.lemmatize(word) for word in txt.split(' ') if wnl.lemmatize(word) not in stop_words and len(word)>2])
    return txt

data=data.apply(cleaning)
data.shape
```

####  We are going to vectorize the text with a simple approach tfidf Vectorizer to find the most similar vectors (most similar products)

```python
vectorizer=TfidfVectorizer(lowercase=True,stop_words='english')
vectors=vectorizer.fit_transform(data)
vectors.shape
```

```python
#printing the samples of the tfidf vectors for each product 
pd.DataFrame(vectors.todense(), columns=vectorizer.get_feature_names_out(),
            index=df.product_id).sample(5, axis=1).sample(10,axis=0)
```

#### Dimensionality reduction
There is a high amount of tfidf vectors so we are going to apply Principal Component Analysis to reduce the dimensionality for computational efficiency and to reduce the features that are not much informative 

```python
n_components=500
pca=PCA(n_components=n_components)
pca.fit(np.asarray(vectors.todense()))
data2=pca.transform(np.asarray(vectors.todense()))
```

```python
print(f'% of variance explained with {n_components} features: {round(np.sum(pca.explained_variance_ratio_),2)*100}%')
print(data2.shape)
data2[0].shape
```

#### 2.  We need to apply a measure of similarity, such as Cosine similarity, Eucledean distance,Manhattan distance etc. We are going to use Cosine similarity 

```python
def get_recommendation(product_idx,k=3):
    '''product_idx: the index of the product we want recommendation for
       k: Number of recommended products '''
    
    x=data2[product_idx,:].reshape(1,-1) #the product we want to get a recommendation for 
    y=np.delete(data2,product_idx,0) # we delete the product we want to recommend the data for 
    data=df['product_id'].drop(product_idx,axis=0).reset_index(drop=True) # we delate the product we want the recommendation for from the original df
    output=pd.concat([data,pd.DataFrame({'cosine_similarity':cosine_similarity(x,y).reshape(y.shape[0])})],
                     axis=1).sort_values(by='cosine_similarity',ascending=False) 
    return output.iloc[:k,:]
```

```python
#recommendation for product with index 0
product_idx=0
recommendation=get_recommendation(product_idx,k=3)
recommendation
```

### 5.2 Collaborative Filtering 

Collaborative filtering on the other use similarities between users and items. Faster to learn as it uses just dot product of matrices.They try to predict the preference of a user to a product based on the preference of similar users.

We will use user_id, product_id and rating columns


#### We need to use an approach to learn embedding matrix or to learn the latent features for users and products

#### - Matrix factorization approach 
descompose a matrix into two lower-rank matrices. We want to convert the sparse matrix into a dense matrix for this, matrix factorization helps to fill the missing ratings. We will use Non-Negative Matrix Factorization approach but there are many others,such as Single Value Decomposition 

```python
# creating the matrix
matrix=df.pivot(index='user_id',columns='product_id',values='rating')
#replacing nan for 0
matrix.replace(np.NaN,0, inplace=True)
```

```python
#performing matrix factorization using Non-Negative Matrix Factorization
model=NMF()
#user latent factors
factors=model.fit_transform(matrix)
#product latent factors
weights=model.components_
```

```python
#dot product of the two matrices ( factors, weights) to
# get the original matrix 
predicted_ratings=np.dot(factors,weights)
predicted_ratings.shape
```

```python
pred_rating=pd.DataFrame(predicted_ratings,columns=list(matrix.columns),
                        index=list(matrix.index))
```

```python
#predicting
user_id=df.loc[0,'user_id']
#columns with the products the customer has not bought yet 
col=matrix.columns[matrix.loc[user_id] == 0].tolist()
pred_rating=pred_rating.loc[:,col]
recommendation=pred_rating.loc[user_id].idxmax(axis=0)
print(f'for user:   {user_id} \n\nthe product recommended is:\n{recommendation}')
```

Above we implemented a basic collaborative filtering. We can see that it is not accurate and the reason of this is that we do not have enough history as our dataset is very small. We can see that the customer that we chose has only one transacction and the recommended product also has been bought once. This lead to have high sparse matrix and the sparsity pattern of the input matrix might affect the performance and effectiveness of the factorization.
